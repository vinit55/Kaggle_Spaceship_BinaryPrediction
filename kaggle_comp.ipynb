```python
def cabinsplit(x):
    if len(str(x).split('/')) < 3:
        return ['Missing','Missing','Missing']
    else:
        return str(x).split('/')
```

```python
def preprocessing(df):
    #Sum of all expenses of passengers
    df['TotalSpent'] = df['RoomService'] + df['FoodCourt'] + df['ShoppingMall'] + df['Spa'] + df['VRDeck']
    
    #Split Cabin info
    df['TempCabin'] = df['Cabin'].apply(lambda x: cabinsplit(x))
    df['Deck'] = df['TempCabin'].apply(lambda x: x[0])
    df['Side'] = df['TempCabin'].apply(lambda x: x[2])
    df.drop(['TempCabin', 'Cabin'], axis=1, inplace=True)
    
    #
    return df
```
